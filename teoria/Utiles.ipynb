# Codigo para importar una imagen desde una url en internet

```python
import cv2
import urllib
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

print "Logo de la Universidad Nacional del Comahue"
req = urllib.urlopen('http://170.210.81.43/VIRTUAL/theme/academi/pix/LOGOUNC.png')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr,-1) # 'load it as it is'
plt.imshow(img)
plt.show()
```
